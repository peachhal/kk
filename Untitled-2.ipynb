```python
import pandas as pd
import matplotlib.pyplot as plt

palette = ['#483D8B', '#8FBC8F', '#800000', '#000080', '#778899']

# Загрузка данных
file_path = 'tate.csv'
df = pd.read_csv(file_path)

df['artist.gender'] = df['artist.gender'].fillna('Not Mentioned')
gender_counts = df['artist.gender'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(gender_counts.index, gender_counts.values, color=palette[0]) #alpha=0.9)

plt.title('Artist Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('gender.png')
#plt.show()
```

```python
# Step 2: Filter artists based on gender and status
# Assuming 'artist.gender', 'artist.death.year' are columns in your CSV
alive_males = df[(df['artist.gender'] == 'Male') & (df['artist.death.year']!= 0)]
dead_males = df[(df['artist.gender'] == 'Male') & (df['artist.death.year'] == 0)]
alive_females = df[(df['artist.gender'] == 'Female') & (df['artist.death.year']!= 0)]
dead_females = df[(df['artist.gender'] == 'Female') & (df['artist.death.year'] == 0)]

# Combine counts for plotting
counts_alive = {'Male': len(alive_males), 'Female': len(alive_females)}
counts_dead = {'Male': len(dead_males), 'Female': len(dead_females)}

# Step 3: Create a bar graph
fig, ax = plt.subplots()
ax.bar(counts_alive.keys(), counts_alive.values(), color=palette[1], label='Alive')
ax.bar(counts_dead.keys(), counts_dead.values(), bottom=list(counts_alive.values()), color=palette[4], label='Dead')

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Artist Gender Count by Alive/Dead Status')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('artist_death_yeat.png')
#plt.show()
```

```python
# Filter out rows where 'metadata.creation decade' is 0
#filtered_df = df[df['metadata.creation decade'] != 0]
filtered_df = df[df['metadata.creation decade'] >= 10]

df['creation_year'] = df['metadata.creation year'].astype(int)

# Group by 'metadata.creation decade' and count the number of works in each decade
decade_counts = filtered_df.groupby('metadata.creation decade').size().reset_index(name='total_works')

# Create a scatter plot using the filtered data
decade_counts.plot(kind='scatter', x='metadata.creation decade', y='total_works', color=palette[3])

# Customize the plot appearance
plt.title('Scatter Plot of Metadata Creation Decade')
plt.xlabel('Decade')
plt.ylabel('Total Number of Works')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.savefig('creation_decade.png')
#plt.show()
```

```python
# Анализ столбца для извлечения основного носителя
# Предполагаем, что основным носителем является первая часть строки перед первой запятой или другим разделителем
df['Primary Medium'] = df[medium_column].str.split(',').str[0]

# Группировка по первичному носителю и подсчет случаев
medium_counts = df['Primary Medium'].value_counts()

# Фильтрация, позволяющая сохранить только верхние N носителей для удобства чтения, и группировка остальных в "Другое"
top_n = 24
top_mediums = medium_counts.nlargest(top_n)
other_mediums_count = medium_counts.iloc[top_n:].sum()

# Добавляем категорию "Другое"
top_mediums['Other'] = other_mediums_count

# Построение  круговой диаграммы
plt.figure(figsize=(10, 8))
top_mediums.plot.pie(autopct='%1.1f%%', startangle=140, cmap='tab20', title='Distribution of Medium in Data')
plt.ylabel('')  # Удаление метки оси Y для улучшения визуального восприятия
plt.show()

```
